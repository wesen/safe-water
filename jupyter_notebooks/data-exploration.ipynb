```python
## Load modules

import pandas as pd
from IPython.display import display, HTML

swdis_data_path = './data/swdis/'
```

## Sampling of all CSV files in swdis.zip

```python
import glob
path = swdis_data_path  + "*.csv"
df_list = []
for fname in glob.glob(path):
    print fname
    
    df = pd.read_csv(fname, error_bad_lines=False, index_col=False)
    df_list += [df]

    display(df.sample(10))
    display(df.nunique().sort_values())
```
